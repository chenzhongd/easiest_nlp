```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn.utils.rnn import pad_sequence
from torch.utils.data import Dataset, DataLoader
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 示例数据集
texts = [
    "I love machine learning",
    "I love deep learning",
    "I love natural language processing",
    "deep learning is fascinating",
]

# 预处理文本数据
class Vocabulary:
    def __init__(self):
        self.word2idx = {}  # 单词到索引的映射
        self.idx2word = {}  # 索引到单词的映射
        self.idx = 0  # 当前可用索引

    def add_word(self, word):
        # 如果单词不在词汇表中，则添加它
        if word not in self.word2idx:
            self.word2idx[word] = self.idx
            self.idx2word[self.idx] = word
            self.idx += 1

    def __len__(self):
        return len(self.word2idx)

    def __getitem__(self, word):
        # 返回单词的索引，如果不在词汇表中则返回 "<UNK>" 的索引
        return self.word2idx.get(word, self.word2idx["<UNK>"])

    def get_word(self, idx):
        # 返回索引对应的单词，如果索引无效则返回 "<UNK>"
        return self.idx2word.get(idx, "<UNK>")
```

```python
# 初始化词汇表并添加特殊标记
vocab = Vocabulary()
vocab.add_word("<PAD>")
vocab.add_word("<UNK>")

# 将所有文本中的单词添加到词汇表中
for text in texts:
    for word in text.split():
        vocab.add_word(word)

# 数据集类
class TextDataset(Dataset):
    def __init__(self, texts, vocab):
        self.vocab = vocab
        self.data = []
        # 将文本数据转换为索引对 (context, target)
        for text in texts:
            words = text.split()
            indices = [vocab[word] for word in words]
            for i in range(1, len(indices)):
                # 使用前 i 个词作为上下文，第 i 个词作为目标
                self.data.append((indices[:i], indices[i]))

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        # 返回上下文和目标词的索引张量
        return torch.tensor(self.data[idx][0]), torch.tensor(self.data[idx][1])

# 填充序列
def collate_fn(batch):
    inputs, targets = zip(*batch)
    # 填充输入序列以使其具有相同的长度
    inputs = pad_sequence(inputs, batch_first=True, padding_value=vocab["<PAD>"])
    targets = torch.stack(targets)  # 将目标词堆叠成一个张量
    return inputs, targets

# 模型定义
class NNLM(nn.Module):
    def __init__(self, vocab_size, embed_size, hidden_size):
        super(NNLM, self).__init__()
        self.embed = nn.Embedding(vocab_size, embed_size)  # 嵌入层
        self.fc1 = nn.Linear(embed_size, hidden_size)  # 隐藏层
        self.fc2 = nn.Linear(hidden_size, vocab_size)  # 输出层

    def forward(self, x):
        x = self.embed(x)  # 将输入的索引转换为嵌入向量
        x = x.mean(dim=1)  # 简单地对嵌入进行平均
        x = torch.relu(self.fc1(x))  # 应用 ReLU 激活函数
        x = self.fc2(x)  # 输出未归一化的得分
        return x
```

```python
# 模型初始化
embed_size = 10
hidden_size = 50
model = NNLM(len(vocab), embed_size, hidden_size)

# 损失函数和优化器
criterion = nn.CrossEntropyLoss()  # 使用交叉熵损失
optimizer = optim.Adam(model.parameters(), lr=0.001)  # 使用 Adam 优化器

# 数据加载器
dataset = TextDataset(texts, vocab)
data_loader = DataLoader(dataset, batch_size=2, collate_fn=collate_fn, shuffle=True)

# 训练模型
epochs = 100
train_losses = []  # 用于存储每个 epoch 的训练损失

for epoch in range(epochs):
    total_loss = 0  # 用于累计每个 epoch 的总损失
    for inputs, targets in data_loader:
        outputs = model(inputs)  # 前向传播
        loss = criterion(outputs, targets)  # 计算损失
        
        optimizer.zero_grad()  # 清除梯度
        loss.backward()  # 反向传播
        optimizer.step()  # 更新权重
        
        total_loss += loss.item()  # 累加损失

    # 计算并记录每个 epoch 的平均损失
    average_loss = total_loss / len(data_loader)
    train_losses.append(average_loss)
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {average_loss:.4f}")

# 绘制损失曲线
plt.figure(figsize=(10, 5))
plt.plot(range(1, epochs + 1), train_losses, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Curve')
plt.legend()
plt.show()

# 测试模型
def predict_next_word(model, vocab, text):
    model.eval()  # 切换到评估模式
    with torch.no_grad():  # 禁用梯度计算
        words = text.split()
        indices = torch.tensor([vocab[word] for word in words]).unsqueeze(0)  # 将输入转换为索引
        output = model(indices)  # 模型预测
        predicted_idx = torch.argmax(output, dim=1).item()  # 获取预测的索引
        return vocab.get_word(predicted_idx)  # 返回预测的单词
```

```python
# 示例预测
input_text = "I love"
predicted_word = predict_next_word(model, vocab, input_text)
print(f"Predicted next word for '{input_text}': {predicted_word}")
```
